```python
import pandas as pd

data = pd.read_csv("/mnt/wsl/data/cleaned_data.csv")

data.head(10)
```

```python
data.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Initialize the encoder
encoder = OrdinalEncoder()

# Fit the encoder and transform the data to ordinal
data['NWCG_GENERAL_CAUSE_Ordinal'] = encoder.fit_transform(data[['NWCG_GENERAL_CAUSE']])

# Create and save the mapping from ordinal encoding
ordinal_mapping = {i: label for i, label in enumerate(encoder.categories_[0])}
print("Ordinal Encoding Mapping:", ordinal_mapping)
```

```python
# Print class distribution in percentages
class_distribution_percent = data['NWCG_GENERAL_CAUSE_Ordinal'].value_counts(normalize=True) * 100
print(class_distribution_percent)
```

```python
from sklearn.model_selection import train_test_split

# Defining the feature set by excluding specific columns
exclude_columns = ['NWCG_CAUSE_CLASSIFICATION','NWCG_GENERAL_CAUSE_Ordinal', 'NWCG_GENERAL_CAUSE','0', 'FPA_ID', 'FOD_ID', 'STATE',
                  'COUNTY', 'STATE_COUNTY', 'GEOID', 'FIRE_YEAR', 'FIRE_SIZE_CLASS', 'DISCOVERY_DOY', 'DISCOVERY_TIME', 'FIRE_SIZE', 'LATITUDE', 'LONGITUDE']
features = [col for col in data.columns if col not in exclude_columns]

# Defining X and y
X = data[features]
y = data[['NWCG_GENERAL_CAUSE_Ordinal']]  

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, roc_auc_score

# Define a dictionary to hold models
models = {
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1),
    "ExtraTrees": ExtraTreesClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1),
    "LogisticRegression (OvR)": OneVsRestClassifier(LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000)),
    "LightGBM": LGBMClassifier(n_estimators=100, random_state=42, class_weight='balanced'),
    "KNN": KNeighborsClassifier(n_jobs=-1),
    "DecisionTrees": DecisionTreeClassifier(random_state=42, class_weight='balanced')
}

# Loop through models, train, predict, and print scores
for name, model in models.items():
    model.fit(X_train, y_train.values.ravel())
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)  # For AUC, we need probabilities
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Calculate ROC AUC score, specifying 'ovr' for multi_class parameter
    auc_score_micro = roc_auc_score(y_test, y_pred_proba, average='micro', multi_class='ovr')  

    print(f"{name} - Accuracy: {accuracy:.4f}, Micro-average AUC (OVR): {auc_score_micro:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l1_l2
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

input_shape = (X_train.shape[1],)  

# Define the model
model = Sequential([
    InputLayer(shape=input_shape),
    Dense(128, activation='gelu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
    BatchNormalization(),
    Dropout(0.2),
    Dense(128, activation='gelu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
    BatchNormalization(),
    Dropout(0.2),
    Dense(13, activation='softmax')  # 13 classes
])
```

```python
from tensorflow.keras.utils import to_categorical

y_train_encoded = to_categorical(y_train, num_classes=13)
```

```python
# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy', 'auc']
)

# Define early stopping and model checkpointing callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)

# Train the model
history = model.fit(
    X_train, y_train_encoded,
    validation_split=0.2,
    epochs=100,  
    callbacks=[early_stopping, model_checkpoint],
    batch_size=32, 
    verbose=1
)
```

```python
from tensorflow.keras.models import load_model
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

# Ensure the test labels are also one-hot encoded
y_test_encoded = to_categorical(y_test, num_classes=13)  

# Load the saved model
best_model = load_model('best_model.keras')

# Predict classes and probabilities on the test set
y_pred_probs = best_model.predict(X_test)
y_pred_classes = np.argmax(y_pred_probs, axis=1)
y_true_classes = np.argmax(y_test_encoded, axis=1)

# Calculate metrics
accuracy = accuracy_score(y_true_classes, y_pred_classes)
auc_micro = roc_auc_score(y_test_encoded, y_pred_probs, average='micro')
auc_macro = roc_auc_score(y_test_encoded, y_pred_probs, average='macro')

print(f"Accuracy: {accuracy}")
print(f"AUC (Micro): {auc_micro}")
print(f"AUC (Macro): {auc_macro}")

# Classification report
print(classification_report(y_true_classes, y_pred_classes))

# Model summary
best_model.summary()

# Plotting training and validation loss
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# If you tracked validation accuracy, you can plot it like this:
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
